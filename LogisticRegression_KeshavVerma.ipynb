importing the required libraries and modules

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

I have uploaded the dataset on google drive and made it public and thereafter fetched the dataset in colab notebook using the following method

```python
# https://drive.google.com/file/d/1mGncv0_QVN3H0v-uQRK8NxpqC2ybg11h/view?usp=sharing

# Replacing FILE_ID with the actual ID from your link
# The FILE_ID is the part between /d/ and /view

file_id = "1mGncv0_QVN3H0v-uQRK8NxpqC2ybg11h"
file_url = f"https://drive.google.com/uc?id={file_id}"
```

loading the dataset

```python
dataset = pd.read_csv(file_url)
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

printing the structure of the dataset and first few entries of it

```python
dataset.info() # printing the summary of the dataframe
```

```python
dataset.head(10)
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f}".format(accuracies.std()*100))
```
