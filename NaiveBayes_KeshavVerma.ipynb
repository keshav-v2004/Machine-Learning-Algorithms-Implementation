## Naive Bayes ML Model on Email Spam dataset from Kaggle by Keshav Verma

importing the required libraries and modules

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

I have uploaded the dataset on google drive and made it public and thereafter fetched the dataset in colab notebook using the following method

```python
# https://drive.google.com/file/d/1Xvd-3OqJxfMpCwKW86N55xyxcZ0_DYu-/view?usp=sharing

# Replacing FILE_ID with the actual ID from your link
# The FILE_ID is the part between /d/ and /view

file_id = "1Xvd-3OqJxfMpCwKW86N55xyxcZ0_DYu-"
file_url = f"https://drive.google.com/uc?id={file_id}"
```

Loading the dataset and thereafter Data Preprocessing

```python
# Load the dataset
df = pd.read_csv(file_url, encoding="latin-1")

# Preprocess the data
df = df[['v1', 'v2']].rename(columns={'v1': 'label', 'v2': 'message'})
df['label'] = df['label'].map({'ham': 0, 'spam': 1})  # Convert labels to numerical values
```

printing dataset information(structure) and first few entries of the dataset

```python
df.info() # printing the summary of the dataframe
```

```python
df.head(10)
```

splitting the dataset into testing and training and then data encoding

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)

# Convert text to numerical features using CountVectorizer
vectorizer = CountVectorizer(stop_words='english')
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)
```

model training and making predictions

```python
# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_vect, y_train)

# Make predictions
y_pred = model.predict(X_test_vect)
```

evaluating the model and printing the different metrics

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy:.4f}')
print('Classification Report:\n', classification_rep)
```

```python
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Confusion Matrix: \n", cm, "\n")
print("Classification Report: \n", report)
print("Accuracy: ", accuracy)
```
