## Ensemble Learning using Bagging technique and using Email Spam dataset from Kaggle by Keshav Verma

importing the required libraries and modules

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import AdaBoostClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

I have uploaded the dataset on google drive and made it public and thereafter fetched the dataset in colab notebook using the following method

```python
# https://drive.google.com/file/d/1Xvd-3OqJxfMpCwKW86N55xyxcZ0_DYu-/view?usp=sharing

# Replacing FILE_ID with the actual ID from our link
# The FILE_ID is the part between /d/ and /view

file_id = "1Xvd-3OqJxfMpCwKW86N55xyxcZ0_DYu-"
file_url = f"https://drive.google.com/uc?id={file_id}"
```

loading the datset and printing dataset info and first few records

```python
# Load the dataset
df = pd.read_csv(file_url, encoding='latin-1')

# Keep only relevant columns
df = df[['v1', 'v2']].rename(columns={'v1': 'label', 'v2': 'message'})
```

```python
df.info() # printing the summary of the dataframe
```

```python
df.head(10)
```

data encoding

```python
# Convert labels to binary (spam = 1, ham = 0)
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)
```

```python
# Convert text into TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

creating and training the model

```python
# Create Bagging classifier using Decision Trees as base learners
bagging_model = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, random_state=42)

# Train the model
bagging_model.fit(X_train_tfidf, y_train)
```

prediction and evaluation

```python
# Predict and evaluate
y_pred = bagging_model.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.4f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Confusion Matrix: \n", cm, "\n")
print("Classification Report: \n", report)
print("Accuracy: ", accuracy)
```

example inputs and predictions

```python
# Example inputs and predictions
example_messages = [
    "Congratulations! You've won a free lottery ticket. Call now!",
    "Hey, are we still meeting for lunch today?",
    "URGENT! Your account has been compromised. Click the link to secure it.",
    "Don't forget to submit your assignment by tonight."
]

# Transform the example messages
example_tfidf = vectorizer.transform(example_messages)

# Predict their labels
example_predictions = bagging_model.predict(example_tfidf)

# Print results
for message, prediction in zip(example_messages, example_predictions):
    label = "Spam" if prediction == 1 else "Ham"
    print(f"Message: {message}\nPrediction: {label}\n")
```

## Ensemble Learning using Boosting technique and using Email Spam dataset from kaggle by Keshav Verma 2023UCS1536



loading the datset which was uploaded on google drive into colab notebook and thereafter eliminating irrelevent columns

```python
# Load the dataset
df = pd.read_csv(file_url, encoding='latin-1')

# Keep only relevant columns
df = df[['v1', 'v2']].rename(columns={'v1': 'label', 'v2': 'message'})
```

data encoding and data splitting

```python
# Convert labels to binary (spam = 1, ham = 0)
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)
```

```python
# Convert text into TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

creating and training the classifier(ML Model)

```python
# Create AdaBoost classifier using Decision Trees as base learners
boosting_model = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1), n_estimators=50, random_state=42)

# Train the model
boosting_model.fit(X_train_tfidf, y_train)
```

prediciting and evaluating

```python
# Predict and evaluate
y_pred = boosting_model.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.4f}")
```

sample inputs and their predicted output

```python
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Confusion Matrix: \n", cm, "\n")
print("Classification Report: \n", report)
print("Accuracy: ", accuracy)
```

```python
# Example inputs and predictions
example_messages = [
    "Congratulations! You've won a free lottery ticket. Call now!",
    "Hey, are we still meeting for lunch today?",
    "URGENT! Your account has been compromised. Click the link to secure it.",
    "Don't forget to submit your assignment by tonight."
]

# Transform the example messages
example_tfidf = vectorizer.transform(example_messages)

# Predict their labels
example_predictions = boosting_model.predict(example_tfidf)

# Print results
for message, prediction in zip(example_messages, example_predictions):
    label = "Spam" if prediction == 1 else "Ham"
    print(f"Message: {message}\nPrediction: {label}\n")
```
